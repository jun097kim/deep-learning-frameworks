```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w1 = 1.0
w2 = 1.0
b = 1.0
```

```python
def forward(x):
    return (x * x) * w2 + x * w1 + b

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

def gradient_w1(x, y):
    return 2 * x * (x * w1 - y + w2 * (x * x) + b)

def gradient_w2(x, y):
    return 2 * (x * x) * (x * x * w2 - y + w1 * x + b)

def gradient_b(x, y):
    return 2 * (b - y + w2 * (x * x) + w1 * x)
```

```python
# Before training
print("predict (before training)", 4, forward(4))

# Training loop
for epoch in range(200):
        for x_val, y_val in zip(x_data, y_data):
            grad_w1 = gradient_w1(x_val, y_val)
            grad_w2 = gradient_w2(x_val, y_val)
            grad_b = gradient_w2(x_val, y_val)
            w1 = w1 - 0.01 * grad_w1
            w2 = w2 - 0.01 * grad_w2
            b = b - 0.01 * grad_b
            print("\tgrad: ", x_val, y_val, grad_w1, grad_w2, grad_b)
            l = loss(x_val, y_val)
            
        print("progress:", epoch, "w1=", w1, "w2=", w2, "b=", b, "loss=", l)

# After training
print ("predict (after training)", "4 hours", forward(4))
```
